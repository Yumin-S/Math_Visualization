### 1. The solution of the inequality set.
a1x+b1y+c1z+d1>0
<br> a1x+b1y+c1z+d1>0
<br> a1x+b1y+c1z+d1>0

```python
import numpy as np
import plotly.graph_objects as go

def plot_solution_space(a1, b1, c1, d1, a2, b2, c2, d2, a3, b3, c3, d3):
    
    # Define the range of data points shown in the picture
    # You can change the values as you wish to adjust the scale:
    # x_range = np.linspace(starting, ending, nunmber of points between them)
    x_range = np.linspace(-1000, 1000, 20)
    y_range = np.linspace(-1000, 1000, 20)
    z_range = np.linspace(-1000, 1000, 20)
    
    X, Y, Z = np.meshgrid(x_range, y_range, z_range)
    X, Y, Z = X.flatten(), Y.flatten(), Z.flatten()
    
    # Applying the inequalities
    valid_points = (a1*X + b1*Y + c1*Z + d1 > 0) & \
                   (a2*X + b2*Y + c2*Z + d2 > 0) & \
                   (a3*X + b3*Y + c3*Z + d3 > 0)
    
    X, Y, Z = X[valid_points], Y[valid_points], Z[valid_points]

    # Plotting
    fig = go.Figure(data=[go.Scatter3d(x=X, y=Y, z=Z, mode='markers',marker=dict(size=3))])
    fig.update_layout(title='Solution Space of Inequalities', autosize=False,
                      width=800, height=800,
                      margin=dict(l=65, r=50, b=65, t=90))
    fig.show()
    
## Change the values of parameters as you wish, to look for results for different inequality sets.
plot_solution_space(a1=1, b1=2, c1=3, d1=4, a2=4, b2=3, c2=2, d2=1, a3=9, b3=10, c3=11, d3=12)
```
